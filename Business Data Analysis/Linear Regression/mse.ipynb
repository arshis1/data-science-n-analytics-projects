```python
'''Implementing mathematical formulas that work on matrices and vectors is a key use case to consider
NumPy for. For example, consider the mean square error formula that is central to regression
problems, which is defined as:

                            1  n
                    MSE=    - ∑   (y_prediction -  y)^2
                            n  i=1

Question: You need to define y_prediction to hold the valus [1,3,4.5] and y to hold the values
[1.1, 2.5, 5.5], and using numpy arithmtic and aggregation to compute MSE
'''
```

```python
import numpy as np
y_prediction = np.array([1, 3, 4.5])
y_prediction
```

```python
y = np.array([1.1, 2.5, 5.5])
y
```

```python
mse = np.mean(y_prediction - y) ** 2
mse
```
