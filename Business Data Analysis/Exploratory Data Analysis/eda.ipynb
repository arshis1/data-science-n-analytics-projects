```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
print(iris.DESCR)
```

```python
iris_attr = iris.data
iris_attr
```

```python
species = iris.target
species
```

```python
print(iris.feature_names)
```

```python
#Print out the shape of iris_attr.
print(iris_attr.shape)
```

```python
#2. Print out the 3rd row of iris_attr.
print(iris_attr[2])
```

```python
# 3. Make a new variable, sepal width, containing the entire second column of iris_attr.
sepal_width = iris_attr[:,1]
print(sepal_width)
```

```python
#4.4. What is the petal length (the 3rd column) of the 95th iris plant?

print(iris_attr[94, 2])
```

```python
#5. Find observations whose species are “versicolor” (whose target value is 1).
versicolor = iris_attr[iris.target == 1]
print(versicolor)
```

```python
import numpy as np

#6. Find out the average petal length of all versicolors using np.mean()

petal_length_index = iris.feature_names.index('petal length (cm)')
print(np.mean(versicolor[:, petal_length_index]))
```
